```python
# Reverse the number
num=int(input("Enter the value :"))
reversed_num=""
while(num>0):
    digit=num%10
    reversed_num+=str(digit)
    num=num//10
print(reversed_num)
```

```python
# anagram check
str_1=input("enter the string_1::")
str_2=input("enter the string_2::")
if(sorted(str_1)==sorted(str_2)):
    print("anagram")
else:
    print("not an anagram")
```

```python
# find max and min of 3 numbers
a = 10
b = 25
c = 7

# assume first value
max_val = a
min_val = a

# compare with b
if b > max_val:
    max_val = b
if b < min_val:
    min_val = b

# compare with c
if c > max_val:
    max_val = c
if c < min_val:
    min_val = c

print("Max:", max_val)
print("Min:", min_val)
```

```python
#sum of n natural numbers

n = int(input("Enter the integer value: "))
total = 0

for i in range(1, n+1):
    total = total + i
    print("Sum after adding", i, "is", total)

print("Final sum:", total)
```

```python
#add multiple numbers
nums = list(map(int, input("Enter numbers: ").split()))

total = 0
for num in nums:
    total += num

print("Sum:", total)
```
