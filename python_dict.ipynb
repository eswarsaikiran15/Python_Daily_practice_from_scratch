# Dictionary in Python

## Definition

A **dictionary** in Python is a collection of **key-value pairs**.

* Keys map to values
* Access is done using **keys (not index)**
* It is an **unordered collection** *(in modern Python, insertion order is preserved, but conceptually unordered)*

---

## Important Points

* Keys must be **unique and immutable** (`int`, `str`, `tuple`)
* Values can be of **any data type**
* Duplicate keys ❌ not allowed
* Duplicate values ✅ allowed
* Supports **dictionary comprehension**
* Think of:

  * **Keys → Roll numbers**
  * **Values → Student names**

---

## Example

```python
d = {"a": 1, "b": 2}
print(d["a"])  # Output: 1
```

---

## Most Used Dictionary Functions and Methods

| Function / Method | Explanation                                             |
| ----------------- | ------------------------------------------------------- |
| `dict.keys()`     | Returns all keys in the dictionary                      |
| `dict.values()`   | Returns all values in the dictionary                    |
| `dict.items()`    | Returns key-value pairs as tuples                       |
| `dict.pop(key)`   | Removes and returns value of given key                  |
| `dict()`          | Creates a dictionary or converts iterable to dictionary |
| `dict.get(key)`   | Returns value of key (no error if missing)              |
| `dict.update()`   | Adds/updates key-value pairs                            |

---

## Example Usage

```python
d = {"a": 1, "b": 2}

print(d.keys())    # dict_keys(['a', 'b'])
print(d.values())  # dict_values([1, 2])
print(d.items())   # dict_items([('a', 1), ('b', 2)])

print(d.get("a"))  # 1

d.update({"c": 3})
print(d)           # {'a': 1, 'b': 2, 'c': 3}

d.pop("b")
print(d)           # {'a': 1, 'c': 3}
```

---

## When to Use Dictionary

### Use when:

* You need **key-value mapping**
* Fast lookup (`O(1)` using hashing)
* Structured data storage

### Avoid when:

* Order-based operations are critical
* You only need a simple list

---

## Key Insight

A dictionary is built on **hashing**, just like sets.

* **dict → key : value**
* **set → only keys**

Understanding this is critical:

> Dictionary = HashMap
> Set = HashMap without values

---


````python
# Creating a dictionary
country_capitals = {
    "Germany": "Berlin",
    "Canada": "Ottawa",
    "England": "London"
}

# Adding an item
country_capitals["Italy"] = "Rome"
print("After adding Italy:", country_capitals)

# Accessing values and keys
print(country_capitals.values())
print(country_capitals.keys())

# Iterating through the dictionary
print("\nIterating through the dictionary:")
for country, capital in country_capitals.items():
    print(f"{country}: {capital}")

# Removing an item using pop()
capital = country_capitals.pop("Canada")
print("\nAfter popping Canada (removed capital):", capital)
print("Current dictionary:", country_capitals)

# Clearing the dictionary
country_capitals.clear()
print("\nAfter clearing the dictionary:", country_capitals)

# Re-creating the dictionary
country_capitals = {
    "Germany": "Berlin",
    "Canada": "Ottawa",
    "England": "London"
}

# Using get() method
print("\nUsing get():", country_capitals.get("Germany"))

# Using update() method
country_capitals.update({"India": "New Delhi", "Japan": "Tokyo"})
print("After update:", country_capitals)

print("key value parirs:",list(country_capitals.items()))
````
