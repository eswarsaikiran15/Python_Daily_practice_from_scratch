```python
# reverse the list (brute force approach)
def reverse_list(input_list):
    list_2=[]
    for i in range(len(input_list)-1,-1,-1):
        list_2.append(input_list[i])
    return list_2

input_list=[1,2,3,4]
print(reverse_list(input_list))
```

```python
# reverse the list (brute force approach)
def reverse_list(input_list):
    list_2=[]
    for i in range(0,len(input_list)):
        n=len(input_list)-1
        list_2.append(input_list[n-i])
    return list_2

input_list=[1,2,3,4]
print(reverse_list(input_list))
```

```python
#Largest Element
def largest_element(lst):
    max_val=lst[0]
    for num in lst:
        if num>max_val:
            max_val=num
    return max_val
lst=[1,2,4,6,23,6,7,1]
print(largest_element(lst))
```

```python
#smallest Element
def smallest_element(lst):
    min_val=lst[0]
    for num in lst:
        if num<min_val:
            min_val=num
    return min_val
lst=[1,2,4,6,23,6,7,1]
print(smallest_element(lst))
```

```python
#Remove Duplicates
def duplicates(lst):
    result=[]
    for num in lst:
        if num not in result:
            result.append(num)
    return result
lst=[1,2,5,3,6,2,2,3,7,8,32]
print(duplicates(lst))
```

```python
#sum of elements
def sum(lst):
    result=0
    for num in lst:
        result+=num
    return result
lst=[1,2,3,4,5,6,7,8,9]
print(sum(lst))
```

```python
# second largest
def second_largest(lst):
    first=second=float('-inf')
    for num in lst:
        if num>first:
            second=first
            first=num
        elif num>second and num!=first:
            second=num
    return second
lst=[1,2,3,4,5,6,7,8,9]
print(second_largest(lst))
```

```python
# palindrome list

def palindrome(lst):
    return lst==lst[::-1]
lst=[1,2,1]
print(palindrome(lst))
```

```python
# count occurences
def count_occurrences(lst, x):
    count = 0
    for num in lst:
        if num == x:
            count += 1
    return count
lst=[1,2,2,2,2,3,4,5,6,7,8,9]
print(count_occurrences(lst, 2))
```

```python
# missing number
def missing_number(lst, n):
    total = n * (n + 1) // 2
    return total - sum(lst)

lst = [1,2,3,4,6,7,8,9]
print(missing_number(lst, 9))
```

```python
# move zeroes to end
def move_zeros(lst):
    pos = 0  # position to place next non-zero

    for i in range(len(lst)):
        if lst[i] != 0:
            lst[pos], lst[i] = lst[i], lst[pos]
            pos += 1

    return lst
lst = [1,2,0,0,3,0,4,5,6,7,8,9]
print(move_zeros(lst))
```
