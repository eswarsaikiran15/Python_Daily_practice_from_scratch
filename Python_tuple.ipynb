# 📘 Python Tuples – Notes

## 🔹 What is a Tuple?

A **tuple** is a built-in data structure in Python used to store multiple items in a single variable.

* Ordered (maintains insertion order)
* Immutable (cannot be modified after creation)
* Allows duplicate values

---

## 🔹 Key Characteristics

* **Immutable**: Elements cannot be added, removed, or changed after creation
* **Indexed**: Supports accessing elements using index positions
* **Ordered**: Maintains the sequence of elements
* **Heterogeneous**: Can store different data types
* **Hashable (if elements are immutable)**: Can be used as keys in dictionaries

---

## 🔹 Why Use Tuples?

* Ensure **data integrity** by preventing modifications
* Achieve **better performance** compared to lists in certain cases
* Represent **fixed collections of data**
* Use as **dictionary keys** (when elements are immutable)
* Return **multiple values from functions** in a structured way

---

## 🔹 Tuple Creation

* Created using parentheses or by separating values with commas
* A single-element tuple requires a trailing comma
* Parentheses are optional but recommended for clarity

---

## 🔹 Accessing Elements

* Elements are accessed using **indexing**
* Supports both **positive and negative indexing**

---

## 🔹 Tuple Slicing

* Supports slicing to access a range of elements
* Slicing creates a **new tuple**

---

## 🔹 Immutability

* Tuple elements cannot be modified after creation
* Any modification attempt results in an error
* If a tuple contains mutable elements, those inner elements can still be modified

---

## 🔹 Tuple Methods

Tuples provide only two built-in methods:

### 1. `count()`

* Returns the number of occurrences of a specified value

### 2. `index()`

* Returns the index of the first occurrence of a specified value

---

## 🔹 Common Operations

* **Length Calculation**: Determine number of elements
* **Membership Test**: Check if an element exists
* **Concatenation**: Combine tuples
* **Repetition**: Repeat elements multiple times

---

## 🔹 Tuple vs List

| Feature     | Tuple      | List         |
| ----------- | ---------- | ------------ |
| Mutability  | Immutable  | Mutable      |
| Methods     | Few        | Many         |
| Performance | Faster     | Slower       |
| Use Case    | Fixed data | Dynamic data |

---

## 🔹 When to Use Tuples

* When data should remain **constant**
* When **performance** is important
* When using data as **dictionary keys**
* When representing **structured, fixed collections**

---

## 🔹 When Not to Use Tuples

* When data needs frequent updates
* When elements need to be added or removed
* When working with highly dynamic datasets

---

## 📌 Summary

* Tuple = **ordered, immutable collection**
* Ensures **data safety and consistency**
* Limited methods but **efficient and reliable**
* Best suited for **fixed, read-only data structures**

---


# 📘 Tuple Methods and Functions

## 🔹 Tuple Methods (Only 2)

These are defined inside the tuple object.

### 1. `count()`

* Returns the number of occurrences of a specified value

### 2. `index()`

* Returns the index of the first occurrence of a specified value

👉 **Fact:** Tuples have only these two methods because they are immutable.

---

## 🔹 Built-in Functions Applicable to Tuples

These are not tuple methods, but they operate on tuples because tuples are iterable.

### 🔸 Basic Functions

* `len()` → Returns number of elements
* `max()` → Returns maximum value
* `min()` → Returns minimum value
* `sum()` → Returns sum of elements

---

### 🔸 Utility Functions

* `sorted()` → Returns a sorted list from tuple elements
* `all()` → Returns `True` if all elements are true
* `any()` → Returns `True` if at least one element is true

---


```python
# Creating tuples
fruits = ("apple", "banana", "cherry")
numbers = (1, 2, 3)

# Accessing elements
first_fruit = fruits[0]      # "apple"
last_fruit = fruits[-1]      # "cherry"

# Tuple length
length_of_fruits = len(fruits)   # 3

# Concatenation
combined_tuple = fruits + ("orange",)
# ("apple", "banana", "cherry", "orange")

# Repetition
repeated_numbers = numbers * 2
# (1, 2, 3, 1, 2, 3)

# Slicing tuples
slice_of_fruits = fruits[1:3]
# ("banana", "cherry")

# Using tuple method
print(numbers.index(2))

# Output results
print("First Fruit:", first_fruit)
print("Last Fruit:", last_fruit)
print("Length of Fruits:", length_of_fruits)
print("Combined Tuple:", combined_tuple)
print("Repeated Numbers:", repeated_numbers)
print("Slice of Fruits:", slice_of_fruits)
```

```python
# Note:
# Tuples are immutable, meaning we cannot add elements directly.
# To "add" an element, we create a new tuple by concatenating
# the existing tuple with a single-element tuple.

# Important:
# (new_value,) → the comma is required to create a single-element tuple.
# Without the comma, it is treated as a normal value, not a tuple.

# Also:
# This operation does NOT modify the original tuple.
# It creates a new tuple and reassigns it to the variable.
n=4
tuple=(10,40)
for i in range(0,n):
    new_value=int(input("enter the tuple:"))
    tuple = tuple + (new_value,) # why comma mentioned means it created single element tuple
    print(tuple)


for i in range(0,len(tuple)):
    print(tuple[i])


```

```python
#find the index of an element in the tuple
tuple=(10,30,20,40,70)
find=20
for i in range(0,len(tuple)):
    if(find == tuple[i]):
        print("value is found at index",i)
        break
```

```python
# count occurences in tuple
# if we perform addition problems we mention count=0 and if suppose it is multiplication then we put count=1
def count_occurrences(tpl, x):
    count = 0
    for num in tpl:
        if num == x:
            count += 1
    return count

lst = [1,2,2,2,2,3,4,5,6,7,8,2,9]

print(count_occurrences(lst, 2))
```
