# Python List Data Structure

## Definition

A **list** in Python is a built-in data structure used to store multiple values in a single variable.
It is **ordered**, **mutable (changeable)**, and can hold **different types of data**.

---

## What a List Actually Is

A list behaves like a **dynamic array**:

* It grows and shrinks automatically
* It stores references to objects (not raw values)
* It allows mixed data types

Example:

```python
a = [10, "hello", 3.5]
```

---

## Key Characteristics

* Ordered (maintains insertion order)
* Mutable (can modify elements)
* Allows duplicate values
* Supports indexing and slicing

---

## Creating a List

```python
a = [1, 2, 3]
b = []                # empty list
c = list((4, 5, 6))   # using constructor
```

---

## Important List Methods

### Adding Elements

```python
a = [1, 2]

a.append(3)        # [1, 2, 3]
a.insert(1, 10)    # [1, 10, 2, 3]
a.extend([4, 5])   # [1, 10, 2, 3, 4, 5]
```

---

### Removing Elements

```python
a = [1, 2, 3, 2]

a.remove(2)   # removes first occurrence
a.pop()       # removes last element
a.pop(1)      # removes element at index 1
del a[0]      # deletes element at index 0
```

---

### Searching and Counting

```python
a = [1, 2, 2, 3]

a.index(2)    # returns index of first occurrence
a.count(2)    # returns number of occurrences
```

---

### Sorting and Reversing

```python
a = [4, 2, 1, 3]

a.sort()                # ascending order
a.sort(reverse=True)    # descending
a.reverse()             # reverse list
```

---

### Utility Methods

```python
a = [1, 2, 3]

len(a)        # length of list
a.clear()     # removes all elements
b = a.copy()  # creates a copy
```

---

## Indexing Example

```python
a = [10, 20, 30, 40]

a[0]   # 10
a[-1]  # 40
```

---

## Slicing Example

```python
a = [1, 2, 3, 4, 5]

a[1:4]   # [2, 3, 4]
a[:3]    # [1, 2, 3]
a[::2]   # [1, 3, 5]
```

---

## Summary

A Python list is a flexible, dynamic container used to store and manage multiple values efficiently. It is one of the most commonly used data structures in Python and forms the base for many problem-solving approaches.

---


````python
# 1. Creating a List
my_list = [10, 20, 30, 40]
print("Original List:", my_list)


# 2. Adding Elements

# Append a single element
my_list.append(60)
print("After appending 60:", my_list)

# Insert an element at a specific position
my_list.insert(2, 25)   # Insert 25 at index 2
print("After inserting 25 at index 2:", my_list)

# Extend the list with multiple elements
my_list.extend([70, 80])
print("After extending with [70, 80]:", my_list)


# 3. Removing Elements

# Remove an element by value
my_list.remove(30)
print("After removing 30:", my_list)

# Remove an element by index and return it
removed_element = my_list.pop(1)
print("After popping index 1 (removed element is {}):".format(removed_element), my_list)


# 4. Other Operations

# Length of list
print("Length of list:", len(my_list))

# Count occurrences
print("Count of 20:", my_list.count(20))

# Find index
print("Index of 25:", my_list.index(25))


# 5. Sorting and Reversing

# Sort list
my_list.sort()
print("Sorted list:", my_list)

# Reverse list
my_list.reverse()
print("Reversed list:", my_list)


# 6. Copying List
copy_list = my_list.copy()
print("Copied list:", copy_list)


# 7. Clearing List
my_list.clear()
print("List after clearing:", my_list)
````
